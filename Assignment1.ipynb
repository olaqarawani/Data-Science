**Q3) Create a new column called unified_price that represents the price in USD per kilogram for commodities measured by weight and the price per liter for commodities measured by volume.**

```python
def unified_price(n):
    price = n['usdprice']

    if 'KG' in n['unit']:
        return f"{price}/kg"
    elif 'L' in n['unit']:
        return f"{price}/L"
    elif 'G' in n['unit']:
        return f"{price * 1000}/kg"
    elif 'Cubic meter' in n['unit']:
        return f"{price / 1000}/L"

    return price

df['unified_price'] = df.apply(unified_price, axis=1)
df.head(20)
```

In this question when the unit was 'G' I converted it to kg by multiplying by 1000, and when it was 'Cubic meter' I converted it to L by dividing by 1000
